```python
#checking the unique value_count in each column
for col in df.columns:
    print(df[col].value_counts())
    print("*"*20)
    print("*"*20)
```

```python
#removing spaces from the ADDRESS and change the address as Other for vlaue_count<=10
df['ADDRESS'] = df['ADDRESS'].apply(lambda x:x.strip())
add_count = df['ADDRESS'].value_counts()
```

```python
#vlue of Address having value_count <=10
add_counts10 = add_count[add_count <=10]
add_counts10
```

```python
#changing the Address as other
df['ADDRESS'] = df['ADDRESS'].apply(lambda x: 'other' if x in add_counts10 else x)
```

```python
df['ADDRESS'].value_counts()
```

```python
#removing the value which value < 3000 square_ft/bhk_no.
#because it is outlier
df = df[((df['SQUARE_FT']/df['BHK_NO.'])>=300)]
```
